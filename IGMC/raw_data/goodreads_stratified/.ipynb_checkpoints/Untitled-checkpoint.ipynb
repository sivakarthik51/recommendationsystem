```python
srttrain = train[['user_id', 'book_id']].sort_values(by=['user_id', 'book_id'])
srttest = test[['user_id', 'book_id']].sort_values(by=['user_id', 'book_id'])

srtratingsout = ratingsout[['user', 'item', 'rating']].sort_values(by=['user', 'item'])
ratingstrain = []
i=0
for val in srtratingsout.iterrows():
#     print(val[1])
#     print(srt.iloc[i]['user']==val[1]['user'])
    if (srttrain.iloc[i]['user_id']==val[1]['user'] and  srttrain.iloc[i]['book_id']==val[1]['item']  ):
        ratingstrain.append(val[1]['rating'])
        i+=1

ratingstest = []
i=0
for val in srtratingsout.iterrows():
    if i < len(srttest) and (srttest.iloc[i]['user_id']==val[1]['user'] and  srttest.iloc[i]['book_id']==val[1]['item']  ):
        ratingstest.append(val[1]['rating'])
        i+=1
```

```python
len(ratingstrain), train.shape
```

```python
len(ratingstest), test.shape
```

```python
srttrain['rating'] = ratingstrain
```

```python
srttest['rating'] = ratingstest
```

```python
train
```

```python
train = srttrain.sort_index()
test = srttest.sort_index()
```
